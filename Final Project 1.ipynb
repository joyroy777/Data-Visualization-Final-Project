```python
import pandas as pd
import matplotlib.pyplot as plt

import warnings
```

```python
df = pd.read_csv("tely10.csv")
```

```python
df["city"]=df["city"].str.lower()
```

```python
df
```

```python
john=df[df["city"].str.contains("john")]
```

```python
john
```

```python
remaining=pd.concat([df, john,john ]).drop_duplicates(keep=False)
```

```python
remaining
```

```python
john["city"]="st john's"
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
john
```

```python
result=pd.concat([john,remaining])
```

```python
result
```

```python
bigThree=result[(result["city"]=="st john's")|(result["city"]=="paradise")|(result["city"]=="mount pearl")]
```

```python
others=pd.concat([result, bigThree,bigThree ]).drop_duplicates(keep=False)
```

```python
others
```

```python
others["city"]="other"
```

```python
others
```

```python
done = pd.concat([bigThree,others])
```

```python
final=done.groupby(["year", "city"]).count().reset_index()
```

```python
final
```

```python
almostThere  = final[["year","city","index"]]
```

```python
almostThere
```

```python
almostThere=almostThere.rename(columns={"index":"partcipants"})
```

```python
almostThere
```

```python
graph=almostThere.pivot_table(values="partcipants",index="year",columns="city",aggfunc="first")
```

```python
graph.plot(kind="barh",stacked=True)
```
