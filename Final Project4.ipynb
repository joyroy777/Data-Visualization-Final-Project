```python
chip_time = list(df['chip_time'])
```

```python
def get_seconds(time_str):
    # split in hh, mm, ss
    result =  time_str.split(':')
    if (len(result) == 2):
        mm, ss = time_str.split(':')
        hh = 0
    else:        
        hh, mm, ss = time_str.split(':')
    return int(hh) * 3600 + int(mm) * 60 + int(ss)
```

```python
itemList = []
n = 0
for i in chip_time:
    itemList.append(get_seconds(i))
df["chip_time_in_seconds"] = itemList
```
